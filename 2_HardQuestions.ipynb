```python
def max_subarray_sum(input_arr):
  end = 0
  n = len(input_arr)
  max_sum = 0
  curr_sum = 0
  while end <n:
    if input_arr[end] + curr_sum < 1:
      curr_sum = 0
      end = end + 1
    else:
      curr_sum = curr_sum + input_arr[end]
      max_sum = max(max_sum,curr_sum)
      end = end +1
      
  return max_sum

max_subarray_sum([-1, -3, 5, -4, 3, -6, 9, 2])
```

```python
def has_merge_conflict(pull_requests)-> bool:
  n = len(pull_requests)
  def check(x,y,a,b):
    if y < a:
      return False
    elif x> b:
      return False
    
    else:
      return True

  for i in range(0,n-1):
    for j in range (1,n):
      x,y = pull_requests[i]
      a,b = pull_requests[j]
      if check(x,y,a,b) == True:
        print(x,y,a,b,i,j)
        return True
  

    
    
  return False
```
